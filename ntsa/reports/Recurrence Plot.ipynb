## White Noise

```python
# generating serie

from ntsa.algorithms.series_generator import WhiteNoiseGenerator

white_noite_generator = WhiteNoiseGenerator()

xs = white_noite_generator.generate_series(n=400)
```

```python
# ploting serie
plt.plot(xs)
```

```python
# showing recurrence plot
DataViz.display_recurrence_plot(xs, threshold=0.2)
```

## Periodic Oscilator

```python
# generating serie

from ntsa.algorithms.series_generator import PeriodicOscilatorGenerator

periodic_oscilator_generator = PeriodicOscilatorGenerator(mode="sin")

xs = periodic_oscilator_generator.generate_series(n=400)
```

```python
# ploting serie
plt.plot(xs)
```

```python
# showing recurrence plot
DataViz.display_recurrence_plot(xs, threshold=0.4)
```

## Drift Logistic Map

```python
# generating serie

from ntsa.algorithms.series_generator import DriftLogisticMapGenerator

drift_logistic_map_generator = DriftLogisticMapGenerator(start_point=0.1, r=4, c=0.01)

xs = drift_logistic_map_generator.generate_series(n=150)
```

```python
# ploting serie
plt.plot(xs)
```

```python
# showing recurrence plot
DataViz.display_recurrence_plot(xs, threshold=0.2)
```

## Brownian motion

```python
# generating serie

from ntsa.algorithms.series_generator import BrownianMotionGenerator

brownian_motion_generator = BrownianMotionGenerator(start_point=0, delta=2, dt=0.02)

xs = brownian_motion_generator.generate_series(n=150)
```

```python
# ploting serie
plt.plot(xs)
```

```python
# showing recurrence plot
DataViz.display_recurrence_plot(xs, threshold=0.4)
```
